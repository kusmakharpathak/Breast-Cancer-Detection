```python
#Using Pearson Correlation
plt.figure(figsize=(20,20))
breast_cancer_corr = X.corr()
sns.heatmap(breast_cancer_corr, annot=True, cmap=plt.cm.Reds)
# plt.savefig('E:\Heatmap.png', dpi=1000)
plt.show()
```

```python
high_correlation = breast_cancer_corr.abs()
high_correlation_unstack = high_correlation.unstack()
high_correlation_sort = high_correlation_unstack.sort_values(ascending=False)
print(high_correlation_sort)
print("=============================================================================")
for i in range(len(high_correlation_sort)):
    if high_correlation_sort[i]>0.99 and high_correlation_sort[i]<1:
        print(i)
print("=============================================================================")
print(high_correlation_unstack[1:6
                              ])
print("=============================================================================")
print(high_correlation_sort[30:35])
print("=============================================================================")
```

```python
x_train, x_test, y_train, y_test = train_test_split(final_features.drop('diagnosis', axis=1), final_features['diagnosis'], test_size = 0.30)
```

```python
type(x_train)
```

```python
type(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
logmodel = LogisticRegression(C=100)
```

```python
logmodel.fit(x_train, y_train)
```

```python
prediction = logmodel.predict(x_test)
```

```python
accuracy_score(y_test, prediction)
```

```python
M = pd.DataFrame(final_features[final_features['diagnosis']=="M"])
B = pd.DataFrame(final_features[final_features['diagnosis']=="B"])
```

```python
min_value_false = list(M.min())
min_value_false = min_value_false[:-1]
max_value_false = list(M.max())
max_value_false = max_value_false[:-1]
min_value_true = list(B.min())
min_value_true = min_value_true[:-1]
max_value_true = list(B.max())
max_value_true = max_value_true[:-1]
la = list(final_features.columns)
la = la[:-1]
```

```python
print("Min Value of 0")
print("================================")
print(M.min())
print(min_value_false)
print("================================")
print("Min Value of 1")
print("================================")
print(B.min())
print(min_value_true)
print("================================")
print("Max Value of 0")
print("================================")
print(M.max())
print(max_value_false)
print("================================")
print("Max Value of 1")
print("================================")
print(B.max())
print(max_value_true)
print("================================")
print(la)
```

```python
plt.figure(figsize=(100, 70))
plt.xlabel("Attribute")
plt.ylabel("Mean Value")
plt.title("Breast cancer Detection Range Value Graph")
plt.scatter(la, min_value_false, s = 500)
plt.plot(la, min_value_false, 'r', ls='solid', label='Min = 1', linewidth=10.0)
plt.scatter(la, max_value_false, s = 500)
plt.plot(la, max_value_false, 'g', label='Max = 1', linewidth=10.0)
plt.xlabel("Features", fontsize=100)
plt.ylabel("Report", fontsize=100)
plt.xticks(fontsize=100)
plt.yticks(fontsize=100)
plt.legend()
plt.grid()
# plt.savefig('E:\Range.png', dpi =300)
plt.show()
```
